### Text Mining vs NLP
- Texxt Mining : 비정형 데이터 (텍스트)에서 의미 있는 정보를 추출하는 방법
- NLP(자연어 처리) : 사람이 사용하는 언어를 이해하고 예측하는 분야

#### Text Mining 분랴
- 문서 분류 : Text가 특정 카테고리에 속하는지 예측하는 분야
- 문서 요약 : Text의 주요 내용을 추리는 분야
- 문서 군집분석 : 문서끼지의 유사도를 비교 분석하는 분야
- 감성 분석 : text에 들어있는 감정이나 의도를 분석하는 분야

#### Text Minig 프로세스
1. 텍스트 전처리
   - 클렌징(대/소문자, 특수문자 제거, 오탈자 제거 등)
   - 불용어 제거 (stop words)
   - 어근 추출 (Stemming Lemmatization)
     - 영어 : NLTK
     - 한국어 : Konlpy (형태소 분석기)
   - 토큰화 : text를 일정 단위 (ex.띄어쓰기, 형태소, 2-gram 등)로 분해하는 작업
2. 수치화/Feature Vector화 : 토큰화 된 토큰들을 가지고 일정 규칙에 따라 숫자를 변경하는 작업
   - 빈도 기반의 수치화 (BOW, TF-IDF 등)
   - Word Embedding
3. ML,DL 학습 -> 예측, 평가,분석

### 데이터 클렌징
- 한 글자 단어 삭제 -> 텍스트 마이닝에서는 의미부여가 어려움
- ㅋㅋㅋ, ㅎㅎㅎ 같은 반복적인 글자 삭제

````python
import re # 정규 표현식, Regular Expression
````

````python
# 패턴 생성
p = re.compile("[ㅋㅎㄷ~ㅡ>!]+") # 요소가 두 번 이상 반복되는 경우 패턴을 의미함,, 정규표현식 패턴!
````

```python
# 어떠한 요소가 정규표현식 패턴에 감지되어 있는지 인덱싱 번호 꺼내오기
for i, token in enumerate(인종_국적_token) :
    if 'ㅋㅋ' in token :
        print(i)
```

````python
if p.match(인종_국적_token[48]) :
    print("true였음!")
````

````python
clean_text = []

for token in 인종_국적_token :
    if len(token) < 2 : #글자 개수가 2 미만인가?
        continue
    elif p.match(token) : # 정규 표현식에 매칭되는가?
        continue
    clean_text.append(token) # 나머지 정상 토큰을 리스트에 쌓기
    
````

````python
input_str = " ".join(clean_text)
result_cloud = wc.generate_from_text(input_str)
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(result_cloud)
plt.axis("off")
plt.savefig("./word_cloud.png")
plt.show()
````

### 감성 분석
- 혐오 표현: 0, 악플/욕설 : 1 , clean : 2

````python
train["혐오표현"] = train.loc[:,"여성/가족":"기타 혐오"].sum(axis = 1)

train2 = train[["문장", "혐오표현", "악플/욕설", "clean"]]

train2

def label(row) : #axis를 1로 설정할 경우 학 행들이 Series 형으로 넘어옴
    if row["혐오표현"] >= 1 : 
        return 0
    elif row["악플/욕설"] >= 1 :
        return 1
    else : #clean
        return 2


y_train = train2.apply(label, axis = 1)
y_train
````

````python
# test set 도 정답데이터 만들어주기!
test["혐오표현"] = test.loc[:,"여성/가족":"기타 혐오"].sum(axis = 1)
test2 = test[["문장", "혐오표현", "악플/욕설", "clean"]]

y_test = test2.apply(label, axis = 1)
y_test
````

````python
# 이건 걍 clean 보고 싶어서 내가 확인한 것!
# test2[test2["clean"] == 1]
y_test[y_test[:] == 2]
# train[train["인종/국적"] == 1
````

````python
train_morphs = []
for txt in tq(text_train) :
    # print(txt)  # 시리즈 형인 0차원의 문자열을 보여줌
    morphs = okt.morphs(txt)
    train_morphs.append(morphs)
````

````python
test_morphs = []
for txt in tq(test2["문장"]) :
    # print(txt)  # 시리즈 형인 0차원의 문자열을 보여줌
    morphs = okt.morphs(txt)
    test_morphs.append(morphs)
````

````python
train_morphs
````

````python
p = re.compile("[ㅋㅎㄷ~!.?,0-9a-zA-Z]+")  #정규표현식 패턴 생성
````

````python
clean_text_train = []
for sentence in train_morphs :  # 한 문장씩 꺼내오는 반복문
    # print(sentence) # 2차원의 요소인 1차원을 들여다봄.. 각 단어, 요소에 접근해야 하니까 한 번더 접근 필요
    tmp = []
    for token in sentence :  # 형태소 단위로 쪼개진 token을 꺼내오는 반복문
        # print(token)
        if len(token) < 2:
            continue
        elif p.match(token) :
            continue
        tmp.append(token)
    clean_text_train.append(tmp)
        
````

````python
clean_text_train
````

````python
clean_text_test = []
for sentence in test_morphs :  # 한 문장씩 꺼내오는 반복문
    # print(sentence) # 2차원의 요소인 1차원을 들여다봄.. 각 단어, 요소에 접근해야 하니까 한 번더 접근 필요
    tmp = []
    for token in sentence :  # 형태소 단위로 쪼개진 token을 꺼내오는 반복문
        # print(token)
        if len(token) < 2:
            continue
        elif p.match(token) :
            continue
        tmp.append(token)
    clean_text_test.append(tmp)
        
````

### 수치화(벡터화) 
- (벡터 : 수치로 바뀐 토큰)
- BOW(Bag Of Words) : 문장에서 특정 단어의 등장 빈도수를 수치화하는 방법


````python
from sklearn.feature_extraction.text import CountVectorizer
````

````python
# 벡터화 수행하기 전에 샘플 데이터 생성한 것.
sample_text = [ "나는 오늘 점심에 삼계탕을 먹었다.",
              "너 저녁에 운동갈 거야?",
              "상현아 죽을래?",
              "상현아 미쳤니? 정신 안 차려?..."]
````

````python
sample_cv = CountVectorizer() #BOW 객체 생성(초기화)
````

````python
# 단어 사전 구축

# 띄어쓰기 단위로 토큰화를 수행
# 하나의 단어는 등록하지 않음
sample_cv.fit(sample_text)
````

````python
# 단어 사전 확인
sample_cv.vocabulary_
# 딕셔너리 형태로 제공, 숫자는 컬럼 번호 (인덱스 번호!! 빈도수 아님)
# key : 각각의 단어사전
# value : 컬럽 번호
````

````python
# rsult
result = sample_cv.transform(sample_text)
````
